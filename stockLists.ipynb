```python
"https://pypi.org/project/yfinance/"

import yfinance as yf
import pandas as pd
import json
```

```python
#Get Mylist of tickers
with open('./includes/tickerList.json') as json_file:
    dctStockAll = json.load(json_file)
```

```python
lstWorldTickers = []

for aKey in dctStockAll:
    #dctStockAll[aKey]['history'] = {}
    lstWorldTickers.append(aKey['tickerName'])
    #print(aKey['tickerName'])

#print(dctTickerWorldIndex)

#lstWorldTickers = list(dctStockAll.keys())

tckList = yf.Tickers(lstWorldTickers)
```

```python
# get historical market data
#hist = tckList.history(period="max")
for adDctItem in dctStockAll:
    aDf = pd.DataFrame(tckList.tickers[adDctItem['tickerName']].history(period="max") )
    #strFileName = './results/' + adDctItem['tickerName'] + '.csv'
    #aDf.to_csv(strFileName)
    dctStockAll[adDctItem]["history"] = aDf
```
